2. To find the top 10 major project themes using column 'mjtheme_namecode', let's check what we have in thia column by defalt:

```python
# 'namecode' column
df.mjtheme_namecode.head()
```

This was a nested part in json, which was not parsed to DataFrame properly, so let's use 'json_normalize' to get the data which we need:

```python
# parsing themes from the file using json_normalize
json_string = json.load((open('data/world_bank_projects.json'))) 
themes_df = json_normalize(json_string, 'mjtheme_namecode')
themes_df.head()
```

Here we can see that missing values are represented by empty string. Let's replace it with NaN:

```python
#importing numy which we need for NaN value
import numpy as np

#replacing '' with NaN
themes_df.replace('', np.nan, inplace = True)
themes_df.head()
```

Let's check how many missing and unick values we have in themes_df:

```python
print("'code' missing values count:  " + str(themes_df.code.isnull().sum()))
print("'name' missing values count:  " + str(themes_df.name.isnull().sum()))
```

```python
print("'code' unique values:  " + str(len(themes_df.code.unique())))
print("'name' unique values:  " + str(len(themes_df.name.unique())))
```

'code' has no missing values, while 'name' has 122. 'code' and 'name' has samr number of unique values, taking 'Nan' into consideration.

Now let's check the most popular themes using 'name' column, ignoring missing values for now (we'll get there later).

```python
# most popular themes using 'name'
themes_df.name.value_counts()[:10]
```

3. Now we'll replace missing values using 'code' column. 
First, let's construct dataframe which will serve like a distionary for theme's 'name' and 'code', and then we'll use it to replace missing values.

```python
# constructing dataframe which will serve like a distionary for theme's 'name' and 'code':
d = themes_df.dropna().drop_duplicates().set_index('code')
d
```

```python
#replacing missing values in themes_df using 'code' to 'name' mapping
themes_df.name = themes_df.code.map(lambda x: d.loc[x][0])
themes_df.head()
```

```python
#checking is we have any NaN left
print("'code' missing values count:  " + str(themes_df.code.isnull().sum()))
print("'name' missing values count:  " + str(themes_df.name.isnull().sum()))
```

Now that we fixed missing values, let's re-calculate the most popular themes
